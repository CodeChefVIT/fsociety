```python
import cv2
import numpy
from PIL import Image
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
loc = "Dataset/user_3/"
```

```python
train = pd.read_csv(loc+"user_3_loc.csv")
```

```python
train.head()
```

```python
names = train.image.values.tolist()
```

```python
files = []
for e in names:
    location, file_name = e.split('/')
    files.append(file_name)
```

```python
l = len(files)
```

```python
def Display(image):
    plt.figure(figsize=(14, 10), dpi=80)
    plt.imshow(image)
    plt.show()
```

```python
def showImage(imgname):
    name = loc+imgname
    img = cv2.imread(name)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    Display(img)
```

```python
def findImage(imgname, pos):
    
    redColor = (255, 0, 0)
    lineThickness = 2
    
    name = loc+imgname
    img = cv2.imread(name)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    cv2.rectangle(img, (pos[0], pos[1]), (pos[2], pos[3]), redColor, lineThickness)
    Display(img)
```

```python
def ImageOfObject(num):
    if(num>=0 or num<l):
        pos = train.loc[num].values.tolist()
        location, imgname = pos.pop(0).split('/')
        findImage(imgname,pos)
        #print(pos[0],pos[1],pos[2],pos[3])
    else:
        print("Enter a number betweeen 0-1399")
```
